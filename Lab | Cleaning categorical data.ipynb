```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

2. Load the csv. Use the variable customer_df as customer_df = pd.read_csv().

```python
customer_df = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis2.csv")
customer_df
```

```python
customer_df.columns = customer_df.columns.str.lower()

customer_df.columns = customer_df.columns.str.replace(' ', '_', regex=False)

customer_df = customer_df.rename(columns={
   'employmentstatus': 'employment_status'
})
customer_df
```

```python
#We can drop it
```

```python
customer_df = customer_df.drop(['customer','effective_to_date'], axis=1)
customer_df
```

4. Load the continuous and discrete variables into numericals_df and categorical_df variables, for eg.:
numerical_df = customer_df.select_dtypes()
categorical_df = customer_df.select_dtypes()

```python
numerical_df = customer_df.select_dtypes(['float64', 'int64'])
categorical_df = customer_df.select_dtypes('object')
```

5. Plot every categorical variable. What can you see in the plots?
```python
for col in categorical_df.columns:
    plt.figure(figsize=(10, 4))
    sns.barplot(x=col, y='months_since_policy_inception', data=customer_df, estimator=np.sum)
    plt.title(f'Barplot of Total Number months_since_policy_inception per {col}')
    plt.xticks(rotation=90)  
    plt.show()
```
